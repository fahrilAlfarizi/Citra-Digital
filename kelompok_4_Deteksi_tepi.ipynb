```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Membaca gambar
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar asli
plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
```

```python
# Deteksi tepi menggunakan filter Sobel
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Menghitung magnitudo gradien
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalisasi untuk tampilan yang lebih baik
magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

# Menampilkan hasil deteksi tepi
plt.subplot(2, 3, 2), plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 3, 3), plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#edge manitude (dasaran tepi)
plt.subplot(2, 3, 4), plt.imshow(magnitude, cmap='gray')
plt.title('Edge Magnitude'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Tresholding untuk mendeteksi tepi curam
threshold_value = 50
_, edges_sharp = cv2.threshold(magnitude, threshold_value, 255, cv2.THRESH_BINARY)

# Menampilkan hasil deteksi tepi curam
plt.subplot(2, 3, 5), plt.imshow(edges_sharp, cmap='gray')
plt.title('Sharp Edge'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Tresholding untuk mendeteksi tepi landau
threshold_value = 100
_, edges_ramp = cv2.threshold(magnitude, threshold_value, 255, cv2.THRESH_BINARY)

# Menampilkan hasil deteksi tepi landau
plt.subplot(2, 3, 6), plt.imshow(edges_ramp, cmap='gray')
plt.title('Ramp Edge'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Tresholding untuk mendeteksi tepi garis
threshold_value = 30
_, edges_line = cv2.threshold(magnitude, threshold_value, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 2), plt.imshow(edges_line, cmap='gray')
plt.title('Line Edge'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Tresholding untuk mendeteksi tepi atap
threshold_value = 70
_, edges_roof = cv2.threshold(magnitude, threshold_value, 255, cv2.THRESH_BINARY)

# Menampilkan hasil deteksi tepi atap
plt.figure()
plt.subplot(1, 2, 1), plt.imshow(magnitude, cmap='gray')
plt.title('Roof Edge Magnitude'), plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2), plt.imshow(edges_roof, cmap='gray')
plt.title('Roof Edge'), plt.xticks([]), plt.yticks([])

plt.show()
```
